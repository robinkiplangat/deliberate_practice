[Boston Dataset](http://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_boston.html) Attribute Information:

    1. CRIM      per capita crime rate by town
    2. ZN        proportion of residential land zoned for lots over 
                 25,000 sq.ft.
    3. INDUS     proportion of non-retail business acres per town
    4. CHAS      Charles River dummy variable (= 1 if tract bounds 
                 river; 0 otherwise)
    5. NOX       nitric oxides concentration (parts per 10 million)
    6. RM        average number of rooms per dwelling
    7. AGE       proportion of owner-occupied units built prior to 1940
    8. DIS       weighted distances to five Boston employment centres
    9. RAD       index of accessibility to radial highways
    10. TAX      full-value property-tax rate per USD10,000
    11. PTRATIO  pupil-teacher ratio by town
    12. B        1000(Bk - 0.63)^2 where Bk is the proportion of blacks 
                 by town
    13. LSTAT    % lower status of the population
    14. MEDV     Median value of owner-occupied homes in USD1000's

## Training a Linear Regression Model

* ** First split up our data into an X array(contains the features to train on)**
* **y array with the target variable, in this case the Price column.**


## X and y Arrays

```python
df.columns
```

```python
# X = df[['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']]

X = df.drop('PRICE',axis=1)
```

```python
y = df['PRICE']
```

## Train Test Split with SciKit Learn
split the data into a training set and a testing set. 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

**Instanstiate instance of model**

```python
lm = LinearRegression()
```

**Fit model**

```python
lm.fit(X_train,y_train)
```

## Model Evaluation

**Evaluate the model by check the co-efficients**


```python
lm.intercept_
```

**Coefficient of each feature**

```python
lm.coef_
```

```python
coe_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coe_df
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test, predictions)
```
